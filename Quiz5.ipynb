### โครงการ ที่อยู่ในย่าน (subdistrict) ที่ อาหารการกินมีความหลากหลายที่สุดในช่วงกลางวัน (eating_daytime_variety) ของประเทศไทย

````python
subdistrict = list(set(eat_daytime_variety_['subdistrict_id']))
column_ = list(project.columns)
project_daytime = pd.DataFrame(columns = column_)
for s in subdistrict:
  project0 = project[project['subdistrict_id'] == s]
  project_daytime  = project_daytime.append(project0) 

project_daytime['price_min'] = project_daytime['price_min'].astype(float)
print(project_daytime.shape)
project_daytime
````

ราคาเฉลี่ยของโครงการที่อยู่ในย่านที่ อาหารการกินมีความหลากหลายที่สุดในช่วงกลางวันของประเทศไทย 

````python
avg_proj = project_daytime['price_min'].mean()
print(f'ราคาเฉลี่ยของโครงการ ที่อยู่ในย่าน (subdistrict) ที่ อาหารการกินมีความหลากหลายที่สุดในช่วงกลางวัน (eating_daytime_variety) ของประเทศไทย **{avg_proj:,}** บาท')
print('-'*200)
````

6.1 แสดงย่านที่ อาหารการกินมีความหลากหลายที่สุดในช่วงกลางวัน ของแต่ละจังหวัด

````python
column_names = list(eat.columns)
eat_daytime_sub = pd.DataFrame(columns = column_names)
for p in eat_povince :
   df_p = eat[eat['province_name_th'] == p]
   p_variety = df_p.sort_values(['eating_daytime_variety'],ascending=False).groupby('subdistrict_name_th').head()
   p_variety_df = p_variety.head(1)
   eat_daytime_sub = eat_daytime_sub.append(p_variety_df)
print(eat_daytime_sub.shape)
eat_daytime_sub_ = eat_daytime_sub[['province_name_th' ,'subdistrict_name_th','eating_daytime_variety']]
eat_daytime_sub_ = eat_daytime_sub_.sort_values(['eating_daytime_variety'],ascending=False).reset_index(drop=True)
no = list(eat_daytime_sub_.index+1)
eat_daytime_sub_.insert(0, 'No.', no)
eat_daytime_sub_
````

แสดงย่านที่ อาหารการกินมีความหลากหลายที่สุดในช่วงกลางคืนของแต่ละจังหวัด 

````python
eat_nighttime_sub = pd.DataFrame(columns = column_names)
for p in eat_povince :
   df_p = eat[eat['province_name_th'] == p]
   p_variety = df_p.sort_values(['eating_nighttime_variety'],ascending=False).groupby('subdistrict_name_th').head()
   p_variety_df = p_variety.head(1)
   eat_nighttime_sub = eat_nighttime_sub.append(p_variety_df)
print(eat_nighttime_sub.shape)
eat_nighttime_sub_ = eat_nighttime_sub[['province_name_th' ,'subdistrict_name_th','eating_nighttime_variety']]
eat_nighttime_sub_ = eat_nighttime_sub_.sort_values(['eating_nighttime_variety'],ascending=False).reset_index(drop=True)
no = list(eat_nighttime_sub_.index+1)
eat_nighttime_sub_.insert(0, 'No.', no)
eat_nighttime_sub_
````
