```python
lst_name = list(set(unit_df['propertytype_name_en']))
print(len(lst_name))
lst_name
```

```python
lst_name = list(set(unit_df['propertytype_name_en']))
price_min_len = []
df_all = []
for j in range(len(lst_name)) :
    price_min = np.round(unit_df[unit_df['propertytype_name_en'] == lst_name[j]]['price_min'].mean())
    df_0 = unit_df[unit_df['propertytype_name_en'] == lst_name[j]].fillna({'price_min':price_min})
    price_min_len.append(df_0.shape[0])
    print(f'{lst_name[j]} ---> [DataFrame : Shape] ---> : {df_0.shape}')
    df_all.append(df_0)  
print('-'*100)
## Concat df     
fused_df_name = pd.concat(df_all)
print(fused_df_name.shape)
fused_df_name.head()
```
